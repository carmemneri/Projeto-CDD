```python
import pandas as pd
import plotly.express as px
```

```python
df_perfil_eleitorado = pd.read_csv("perfil_eleitor_secao_2022_PB.csv", encoding="latin1", sep=";")
df_local_votacao = pd.read_csv("eleitorado_local_votacao_2022.csv", encoding="latin1", sep=";")
```

```python
df_perfil_eleitorado = df_perfil_eleitorado.query("NM_MUNICIPIO=='CAMPINA GRANDE'")
df_local_votacao = df_local_votacao.query("NM_MUNICIPIO=='CAMPINA GRANDE'")
```

```python
df_local_votacao.columns
```

```python
df_local_votacao = df_local_votacao[["NR_ZONA", "NR_SECAO", "NR_LATITUDE", "NR_LONGITUDE", "NM_LOCAL_VOTACAO_ORIGINAL", "NM_BAIRRO"]]
df_local_votacao.drop_duplicates(inplace=True)
```

```python
df_perfil_eleitorado.columns
```

```python
df_perfil_eleitorado = df_perfil_eleitorado[['NR_ZONA', 'NR_SECAO', 'DS_GENERO', 'DS_FAIXA_ETARIA', 'DS_GRAU_ESCOLARIDADE', 'QT_ELEITORES_PERFIL']]
df_perfil_eleitorado.drop_duplicates(inplace=True)
```

```python
colunas_comuns = list(filter(lambda x: x in df_perfil_eleitorado.columns, df_local_votacao.columns))
colunas_comuns
```

```python
merge = pd.merge(df_perfil_eleitorado, df_local_votacao, how='outer',
                  left_on=colunas_comuns,
                  right_on=colunas_comuns)

merge.dropna(inplace=True)
```

```python
merge.head()
```

```python
perfis = merge.groupby(["DS_GENERO", "DS_FAIXA_ETARIA", "DS_GRAU_ESCOLARIDADE"]).count().reset_index()[["DS_GENERO", "DS_FAIXA_ETARIA", "DS_GRAU_ESCOLARIDADE"]]
perfis.drop_duplicates(inplace=True)
```

```python
perfis
```

```python
nova_faixa_etaria = {
    "16 anos": "Jovens",
    "17 anos": "Jovens",
    "18 anos": "Jovens",
    "19 anos": "Jovens",
    "20 anos": "Adultos",
    "21 a 24 anos": "Adultos",
    "25 a 29 anos": "Adultos",
    "30 a 34 anos": "Adultos",
    "35 a 39 anos": "Adultos",
    "40 a 44 anos": "Adultos",
    "45 a 49 anos": "Adultos",
    "50 a 54 anos": "Idosos",
    "55 a 59 anos": "Idosos",
    "60 a 64 anos": "Idosos",
    "65 a 69 anos": "Idosos",
    "70 a 74 anos": "Idosos",
    "75 a 79 anos": "Idosos",
    "80 a 84 anos": "Idosos",
    "85 a 89 anos": "Idosos",
    "90 a 94 anos": "Idosos",
    "95 a 99 anos": "Idosos",
    "100 anos ou mais": "Idosos",
    "Inválido": "Inválido"
}
```

```python
perfis["DS_NOVA_FAIXA_ETARIA"] = [nova_faixa_etaria[e.strip()] for e in perfis.DS_FAIXA_ETARIA]
```

```python
perfis = perfis.drop_duplicates(subset=["DS_GENERO", "DS_GRAU_ESCOLARIDADE", "DS_NOVA_FAIXA_ETARIA", "DS_FAIXA_ETARIA"]).query("DS_NOVA_FAIXA_ETARIA!='Inválido'")
perfis
```

```python
perfis_gp = perfis.groupby(["DS_GENERO", "DS_GRAU_ESCOLARIDADE", "DS_NOVA_FAIXA_ETARIA"]).count().reset_index()[["DS_GENERO", "DS_GRAU_ESCOLARIDADE", "DS_NOVA_FAIXA_ETARIA"]]
perfis_gp["NR_PERFIL"] = list(range(perfis_gp.shape[0]))
perfis_gp
```

```python
perfis = pd.merge(perfis, perfis_gp, how="outer", right_on=["DS_GENERO", "DS_GRAU_ESCOLARIDADE", "DS_NOVA_FAIXA_ETARIA"], left_on=["DS_GENERO", "DS_GRAU_ESCOLARIDADE", "DS_NOVA_FAIXA_ETARIA"])
perfis
```

```python
novo_merge = pd.merge(merge, perfis, how='outer',
                  left_on=["DS_GENERO", "DS_FAIXA_ETARIA", "DS_GRAU_ESCOLARIDADE"],
                  right_on=["DS_GENERO", "DS_FAIXA_ETARIA", "DS_GRAU_ESCOLARIDADE"]).dropna()

novo_merge = novo_merge.drop(columns=["DS_FAIXA_ETARIA"]).drop_duplicates()
novo_merge
```

```python
novo_merge.to_csv("PERFIL_LOCALIDADE.csv")
perfis_gp.to_csv("PERFIS.csv")
```
